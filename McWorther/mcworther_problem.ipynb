```python
from mcwhorter import VanGenuchten

model = VanGenuchten()
model.plot()
```

```python
from mcwhorter import McWhorter

problem = McWhorter(model)

problem.plot_D_f()
```

```python
problem.plot_solution()

import numpy as np
import matplotlib.pyplot as plt

# Compare with analytical solution from the online tool of Radek Fučík
# https://mmg.fjfi.cvut.cz/~fucik/index.php?page=exact
# - Parameters are the default values
# - Primary variable is the wetting phase saturation
# - van Genuchten model
# - Original method (McWhorter and Sunada 1990)

# Zeroth column is location, first column is saturation
exact = np.loadtxt("solution_Sw_fucik.txt")

plt.plot(exact[:, 0], exact[:, 1], '--', label="Analytical solution by Radek Fučík")
plt.legend()
plt.show()
```

```python
# with residual saturations.

from mcwhorter import BrooksCorey

model = BrooksCorey(Snr=0.2, Swr=0.3)
model.plot()
problem = McWhorter(model, Si=0.2, S0=0.7)
problem.plot_solution()
```
